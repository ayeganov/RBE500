### Problem 2

a). $\bar{X} = \frac{1}{n}\sum_{i=1}^n x_i$

$ E[\bar{X}] = \frac{1}{n}[E(X_1) + E(X_2) + ... + E(X_n)]$

The $X_i$ are identically distributed, which means they have the same mean $\mu$. All $E(X_i)$ can be replaced by $\mu$.

$ E[\bar{X}] = \frac{1}{n}[\mu + \mu + ... + \mu]$

The sum can be rewritten as $n\mu$ because there are $n$ $\mu$'s.

$ E[\bar{X}] = \frac{1}{n}[n\mu]$

$ E[\bar{X}] = \mu$

b) $Var[\bar{X}] = Var\left(\frac{X_1 + X_2 + ... + X_n}{n}\right)$

$= Var\left(\frac{1}{n}X_1 + \frac{1}{n}X_2 + ... + \frac{1}{n}X_n\right)$

$= \frac{1}{n^2}Var(X_1) + \frac{1}{n^2}Var(X_2) + ... + \frac{1}{n^2}Var(X_n)$

The $X_i$ are identically distributed, which means they have the same variance $\sigma^2$. All $Var(X_i)$ can be replaced by $\sigma^2$.

$= \frac{1}{n^2}\left(\sigma^2 + \sigma^2 + ... + \sigma^2\right)$

The sum can be rewritten as $n\sigma^2$ because there are $n$ $\sigma^2$'s.

### Problem 4

Bayes Formula: $$ P(A \lvert B) = \frac{P(B \vert A)P(A)}{P(B)} $$

Define events: S - small wheel, L - large wheel, B1 - first box, B2 - second box

$$ P(S \lvert B1) = \frac{7}{12}$$

$$ P(S \lvert B2) = \frac{12}{15}$$

$$P(B1) = P(B2) = 0.5$$

Apply bayes formula and Law of Total Probability for P(S):

$$ P(B2 \lvert S) = \frac{P(S \vert B2)P(B2)}{P(S \vert B2)P(B2) + P(S \vert B1)P(B1)} $$

Plug in the numbers:

$$ P(B2 \lvert S) = \frac{\frac{12}{15} * 0.5}{\frac{12}{15} * 0.5 + \frac{7}{12} * 0.5} = 0.5783 $$

```python
import math

import numpy as np
import pandas as pd

master_file = "data/Master.csv"
rover_1_file = "data/Rover_1.csv"
rover_2_file = "data/Rover_2.csv"

rover1 = pd.read_csv(rover_1_file)
rover2 = pd.read_csv(rover_2_file)
master = pd.read_csv(master_file)
 
def gps_to_degrees(gps):
    whole = int(gps / 100)
    minutes = gps - whole * 100
    return whole + minutes / 60.
 
gps_to_degrees = np.vectorize(gps_to_degrees)
 
latitute_rover1_deg = gps_to_degrees(rover1["Latitude"].values)
longitude_rover1_deg = gps_to_degrees(rover1["Longitude"].values)
 
latitute_rover2_deg = gps_to_degrees(rover2["Latitude"].values)
longitude_rover2_deg = gps_to_degrees(rover2["Longitude"].values)
 
latitute_master_deg = gps_to_degrees(master["Latitude"].values)
longitude_master_deg = gps_to_degrees(master["Longitude"].values)

print("PART A")
print("Master mean latitude: {}, mean longitude: {}".format(latitute_master_deg.mean(), longitude_master_deg.mean()))
print("Rover1 mean latitude: {}, mean longitude: {}".format(latitute_rover1_deg.mean(), longitude_rover1_deg.mean()))
print("Rover2 mean latitude: {}, mean longitude: {}".format(latitute_rover2_deg.mean(), longitude_rover2_deg.mean()))


print("\nPART B")

lat_meters_per_deg = 111073.94
long_meters_per_deg = 82805.01


master_lat_errors = (latitute_master_deg - latitute_master_deg.mean()) * lat_meters_per_deg
master_long_errors = (longitude_master_deg - longitude_master_deg.mean()) * long_meters_per_deg
print("Master latitude error std: {}, longitude error std: {}".format(master_lat_errors.std(), master_long_errors.std()))

rover1_lat_errors = (latitute_rover1_deg - latitute_rover1_deg.mean()) * lat_meters_per_deg
rover1_long_errors = (latitute_rover1_deg - latitute_rover1_deg.mean()) * long_meters_per_deg
print("Rover1 latitude error std: {}, longitude error std: {}".format(rover1_lat_errors.std(), rover1_long_errors.std()))

rover2_lat_errors = (latitute_rover2_deg - latitute_rover2_deg.mean()) * lat_meters_per_deg
rover2_long_errors = (latitute_rover2_deg - latitute_rover2_deg.mean()) * long_meters_per_deg
print("Rover2 latitude error std: {}, longitude error std: {}".format(master_lat_errors.std(), master_long_errors.std()))

print("For both rovers and master the latitude errors are larger than the longitude errors")

print("\nPART C")

def distance_between_points(lat1, lon1, lat2, lon2):
    earth_radius = 6371000

    dLat = math.radians(lat2-lat1)
    dLon = math.radians(lon2-lon1)

    lat1 = math.radians(lat1)
    lat2 = math.radians(lat2)

    a = np.sin(dLat/2) * np.sin(dLat/2) + np.sin(dLon/2) * np.sin(dLon/2) * np.cos(lat1) * np.cos(lat2)
    c = 2 * math.atan2(np.sqrt(a), np.sqrt(1-a))
    
    return earth_radius * c;

rover1_to_master = distance_between_points(latitute_rover1_deg.mean(), 
                                           longitude_rover1_deg.mean(), 
                                           latitute_master_deg.mean(),
                                           longitude_master_deg.mean())
rover2_to_master = distance_between_points(latitute_rover2_deg.mean(), 
                                           longitude_rover2_deg.mean(), 
                                           latitute_master_deg.mean(),
                                           longitude_master_deg.mean())

print("Distance between rover1 and master: {} m".format(rover1_to_master))
print("Distance between rover2 and master: {} m".format(rover2_to_master))


print("\nPART D")
rover1_master_matches = rover1.reset_index().merge(master.reset_index(), on=["time"], how="inner", suffixes=["_r1", "_m"])
rover2_master_matches = rover2.reset_index().merge(master.reset_index(), on=["time"], how="inner", suffixes=["_r2", "_m"])

def correct_errors(rover_errors, master_errors, rindex, mindex):
    print("First error: {}".format(rover_errors[11]))

    result = rover_errors.copy()

    for r_i, m_i in zip(rindex, mindex):
        print("ri: {}, mi: {}".format(r_i, m_i))
        print("Copy Value: {}, actual: {}".format(result[r_i], rover_errors[r_i]))
        result[r_i] -= master_errors[m_i]
        break
    return result
        
rover1_lat_errors_corrected = correct_errors(rover1_lat_errors, 
                                             master_lat_errors,
                                             rover1_master_matches["index_r1"], 
                                             rover1_master_matches["index_m"])

print("Rover1 latitude error std: {}, longitude error std: {}".format(rover1_lat_errors_corrected.std(),
                                                                      rover1_long_errors.std()))
```
